```python
import os
import logfire
from dotenv import load_dotenv
from tavily import TavilyClient

load_dotenv(dotenv_path="../.env.local")
logfire.configure()
logfire.instrument_pydantic_ai()


tavily_client = TavilyClient(api_key=os.getenv("TAVILY_API_KEY"))
```

```python
import re

professors_directory = tavily_client.map(
    "https://www.ualberta.ca/en/agricultural-food-nutritional-science/people/faculty.html"
)

# Filter all entries in professors_directory.results that start with the specified string
filtered_entries = [
    entry
    for entry in professors_directory["results"]
    if re.match(r"^https://apps\.ualberta\.ca/directory/person", str(entry))
]
```

```python
import instructor
from openai import OpenAI
from src.types import Lead
from rich import print as rprint


client = instructor.from_openai(OpenAI())

all_leads = []
errors = []
for entry in filtered_entries:
    print(f"Processing {entry}")

    try:
        tavily_response = tavily_client.extract(entry)
        page_content = tavily_response["results"][0]["raw_content"]
        # using instructor to extract the Leads info
        person = client.chat.completions.create(
            model="gpt-4o-mini",
            response_model=Lead,
            messages=[
                {
                    "role": "user",
                    "content": "Extract all the relevant information based on the page content using the lead. The page content is: "
                    + page_content,
                }
            ],
        )
    except Exception as e:
        print(f"Error extracting lead from {entry}: {e}")
        errors.append(entry)
        continue

    all_leads.append(person)


rprint(all_leads)
```
